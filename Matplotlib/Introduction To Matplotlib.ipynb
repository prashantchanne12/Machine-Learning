The pyplot API is generally less-flexible than the object-oriented API. Try to use object-oriented interface over the pyplot interface

```python
# 1st Method
fig = plt.figure() # creates figure
ax = fig.add_subplot() # adds axis
plt.show()
```

```python
# 2nd Method
fig = plt.figure() # creates figure
ax = fig.add_axes([1,1,1,1]) 
ax.plot(x,y) # add some data
plt.show()
```

```python
# 3rd method (recommended)
fig, ax = plt.subplots()
# Create a figure and a set of subplots.
# This utility wrapper makes it convenient to create common layouts of
# subplots, including the enclosing figure object, in a single call.
ax.plot(x,y);
```

## Matplotlib Example Workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [10,20,30,40]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(6,6)) # figsize - size or dimension of the plot/figure (width, height)

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title='Simple Plot',
      xlabel='x-axis',
      ylabel='y-axis')

# 5. save & show (save the whole figure)
fig.savefig('images/sample-plot.png')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [50,60,150,200,250]

fig, ax = plt.subplots(figsize=(6,6))

ax.plot(x,y);

ax.set(title='Sample Graph',
      xlabel='x-axis',
      ylabel='y-axis')

fig.savefig('images/sample-plot2.png')
```

```python
# create some data 
x = np.linspace(0,10,100) # Return evenly spaced numbers over a specified interval.
# in this case 100 different evenly spaces values between 1-10
```

```python
# plot the data
fig,ax = plt.subplots()
ax.plot(x,x**2); # the default plot is called line plot
```

```python
# using same data to create a scatter plot (dots)
fig,ax = plt.subplots()
ax.scatter(x,np.exp(x))
```

#### exp in numpy
The exponential function is e^x where e is a mathematical constant called Euler's number, approximately 2.718281. This value has a close mathematical relationship with pi and the slope of the curve e^x is equal to its value at every point. np.exp() calculates e^x for each value of x in your input array.

```python
fig, ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# Making a plot from dictionary
nut_butter_prices = {'Almond butter':10,
                     'Peanut butter':8,
                     'Cashew butter':12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());

ax.set(title='Butter Price Store',
        xlabel='Types',
        ylabel='Prices in $');
```

## Two Options For Subplots

```python
# Subplots option 1

fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(shoes_sales.keys(),shoes_sales.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,5))

# plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib Objet Oriented)

* When plotting something quickly, okay to use the pyplot method
* When plotting something advanced, use the Object Oriented Method

```python
heart_disease.head()
```

```python
over_fifty = heart_disease[heart_disease['age']>50]
over_fifty
```

```python
len(over_fifty)
```

```python
# pyplot method
over_fifty.plot(kind='scatter',
                x='age',
                y='chol',
                c='target',
                figsize=(10,6)
               )
# c - color
```

```python
# Object Oriented Method mixed with pyplot method

fix,ax = plt.subplots(figsize=(10,6))
over_fifty.plot(kind='scatter',
                x='age',
                y='chol',
                c='target',
                ax=ax);
# plot has the ax param for OO method to plot the data on ax axis

# ax.set_xlim([45,90])
```

```python
# Object Oriented Method From Scratch
fig,ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_fifty['age'],
                     y=over_fifty['chol'],
                     c=over_fifty['target']);

# customize the plot
ax.set(title='Heart Disease and Cholesterol Levels',
       xlabel='Age',
       ylabel='Cholestrol');

# add a legend
ax.legend(*scatter.legend_elements(),title='Target');
# it will automatically find the legend elements from scatter which is c

# add a horizontal line
ax.axhline(over_fifty['chol'].mean(),linestyle='--') # this shows average cholestrol
```

```python
over_fifty.head()
```

```python
# thalach - max heart rate
```

```python
# Subplot of chol, age, thalach

fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10))
# Add data to ax0
scatter = ax0.scatter(x=over_fifty['age'],
                      y=over_fifty['chol'],
                      c=over_fifty['target'])

# customize ax0
ax0.set(title='Heart Disease and Cholesterol Levels',
        xlabel='Age',
        ylabel='Cholestrol')

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title='Target')

# Add a mean line
ax0.axhline(over_fifty['chol'].mean(),
            linestyle='--');


# Add data to ax1
scatter2 = ax1.scatter(x=over_fifty['age'],
                       y=over_fifty['thalach'],
                       c=over_fifty['target'])

# Customizing ax1
ax1.set(title='Heart Disease and Heart Levels',
        xlabel='Age',
        ylabel='Thalach')

# Adding legend
ax1.legend(*scatter2.legend_elements(),title='Target');

# Adding mean line
ax1.axhline(over_fifty['thalach'].mean(),linestyle='--');

# Add a title to the figure
fig.suptitle('Heart Disease Analysis',fontsize=16,fontweight='bold');
```

```python
plt.style.use('seaborn-whitegrid')

# Subplot of chol, age, thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10))
# Add data to ax0
scatter = ax0.scatter(x=over_fifty['age'],
                      y=over_fifty['chol'],
                      c=over_fifty['target'],
                      cmap='winter', # this changes color scheme (color map)
                      s=70)  

# customize ax0
ax0.set(title='Heart Disease and Cholesterol Levels',
        xlabel='Age',
        ylabel='Cholestrol')

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title='Target')

# Add a mean line
ax0.axhline(over_fifty['chol'].mean(),
            linestyle='--');

# changing the x axis limits
ax0.set_xlim([50,80])

# Add data to ax1
scatter2 = ax1.scatter(x=over_fifty['age'],
                       y=over_fifty['thalach'],
                       c=over_fifty['target'],
                       cmap='winter',
                       s=70)

# Customizing ax1
ax1.set(title='Heart Disease and Heart Levels',
        xlabel='Age',
        ylabel='Thalach')

# Adding legend
ax1.legend(*scatter2.legend_elements(),title='Target');

# Adding mean line
ax1.axhline(over_fifty['thalach'].mean(),linestyle='--');

# changing the x axis limits
ax1.set_xlim([50,80])

# changing the y axis limits
ax1.set_ylim([60,200])

# Add a title to the figure
fig.suptitle('Heart Disease Analysis',fontsize=18,fontweight='bold');
```
