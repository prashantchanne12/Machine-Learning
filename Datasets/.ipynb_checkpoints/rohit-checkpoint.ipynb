```python
heart_disease = pd.read_csv("heart-disease.csv")
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
over_50 = heart_disease[heart_disease['age']>50 ]
over_50.head()
```

```python
fix,ax= plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x="age",y="chol",c= "target",ax=ax)
```

```python
over_50.target.values
```

```python
%matplotlib inline
```

```python
fig,ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
ax.set(title="heart disease and cholesterol leveles",
      xlabel="Age",
      ylabel="Cholesterol");
ax.legend(*scatter.legend_elements())
ax.axhline(y=over_50['chol'].mean(),linestyle ='--')
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1,figsize=(10,10))
scatter = ax1.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'])
ax1.set(title="heart disease",xlabel="age",ylabel="cholestrol")
ax1.axhline(y=over_50['chol'].mean(),linestyle='--')
```
